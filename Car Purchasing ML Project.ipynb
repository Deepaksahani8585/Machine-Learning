# -----------Machine Learning---------------------

# 8. Store feature matrix in X and Response(Target) in Vector y

```python
df.columns
```

```python
X = df.drop(['Customer Name', 'Customer e-mail', 'Country','Car Purchase Amount'],axis =1)
```

```python
X
```

```python
y=df['Car Purchase Amount']
```

```python
y
```

# 9.Feature scaling (Normalization)

```python
from sklearn.preprocessing import  MinMaxScaler
```

```python
sc=MinMaxScaler()
X_scaled = sc.fit_transform(X)
```

```python
X_scaled
```

```python
sc1 = MinMaxScaler()
```

```python
y_reshape = y.values.reshape(-1,1)
```

```python
y_scaled = sc1.fit_transform(y_reshape)
```

```python
y_scaled
```

# 10.Splitting the dataset into the training and Test set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_scaled,y_scaled,test_size=0.20,random_state = 42)
```

# 11.Import the models

```python
df.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

# 12.Model Training

```python
lr = LinearRegression()
lr.fit(X_train,y_train)

svm = SVR()
svm.fit(X_train,y_train)

rf = RandomForestRegressor()
rf.fit(X_train,y_train)

gbr = GradientBoostingRegressor()
gbr.fit(X_train,y_train)

xg = XGBRegressor()
xg.fit(X_train,y_train)
```

# 13.Building the ANN

```python
import tensorflow.keras
from keras.models import Sequential
from keras.layers import Dense
```

# 13.1.Initializing the ANN

```python
ann = Sequential()
```

# 13.2 Adding input layer and first hidden layer

```python
ann.add(Dense(25,input_dim=5,activation = 'relu'))
```

# 13.3 Adding second hidden layer

```python
ann.add(Dense(25,activation='relu'))
```

# 13.4 Adding the output layer

```python
ann.add(Dense(1,activation='linear'))
```

# 13.5 Training the ANN

```python
ann.summary()
```

```python
ann.compile(optimizer='adam',loss='mean_squared_error')
```

```python
ann.fit(X_train,y_train,epochs=100,batch_size=50,verbose=1,validation_split=0.2)
```

# 14.Prediction on the test Data

```python
y_pred1 = lr.predict(X_test)
y_pred2 = svm.predict(X_test)
y_pred3 = rf.predict(X_test)
y_pred4 = gbr.predict(X_test)
y_pred5 = xg.predict(X_test)
y_pred6 = ann.predict(X_test)
```

# 15.Evaluating the algorithm

```python
from sklearn import metrics
```

```python
score1 = metrics.r2_score(y_test,y_pred1)
score2 = metrics.r2_score(y_test,y_pred2)
score3 = metrics.r2_score(y_test,y_pred3)
score4 = metrics.r2_score(y_test,y_pred4)
score5 = metrics.r2_score(y_test,y_pred5)
score6 = metrics.r2_score(y_test,y_pred6)
```

```python
print(score1,score2,score3,score4,score5,score6)
```

```python
final_data = pd.DataFrame({'Models':['LR','SVM','RF','GBR','XG','ANN'],
             'R2_Score':[score1,score2,score3,score4,score5,score6]})
```

# 18.GUI

```python
import numpy as np
from tkinter import *
from sklearn.preprocessing import StandardScaler
import joblib
```

```python

def show_entry_fields():
    # Retrieve values from Entry fields
    p1 = float(e1.get())
    p2 = float(e2.get())
    p3 = float(e3.get())
    p4 = float(e4.get())
    p5 = float(e5.get())

    # Load the trained model
    model = joblib.load('car_model')

    # Predict car purchase amount
    result = model.predict(sc.transform(np.array([[p1, p2, p3, p4, p5]])))

    # Display the predicted car purchase amount
    label(master, text='Car Purchase amount').grid(row=8)
    label(master, text=result.inverse_transform(result)).grid(row=10)

    print("Car Purchase amount", result.inverse_transform(result)[0][0])

# Create the main window
master = Tk()
master.title("Car Purchase Amount Predictions Using Machine Learning")

# Create labels and Entry fields
Label(master, text="Gender").grid(row=1)
Label(master, text="Age").grid(row=2)
Label(master, text="Annual Salary").grid(row=3)
Label(master, text="Credit Card Debt").grid(row=4)
Label(master, text="Net Worth").grid(row=5)

e1 = Entry(master)
e2 = Entry(master)
e3 = Entry(master)
e4 = Entry(master)
e5 = Entry(master)

e1.grid(row=1, column=1)
e2.grid(row=2, column=1)
e3.grid(row=3, column=1)
e4.grid(row=4, column=1)
e5.grid(row=5, column=1)

# Create a button to trigger the prediction
Button(master, text='Predict', command=show_entry_fields).grid()

# Start the event loop
mainloop()
```
